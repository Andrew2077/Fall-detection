```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.style.use('seaborn')
```

```python
data_part1 = pd.read_csv('data/data_part1.csv')
data_part2 = pd.read_csv('data/data_part2.csv')
data_part3 = pd.read_csv('data/data_part3.csv')
data_part4 = pd.read_csv('data/data_part4.csv')
```

```python

def plot_data(data, length, with_time):
    fig = plt.figure(figsize=(10, 6))
    if with_time:
        x_axis = data['time']
    else:
        x_axis = data.index.values[0:length]
    plt.plot(x_axis, data['gFx'][0:length], label='gFx')
    plt.plot(x_axis, data['gFy'][0:length], label='gFy')
    plt.plot(x_axis, data['gFz'][0:length], label='gFz')
    plt.legend()
    return plt.show()


```

```python
#* falling 2
plot_data(data_part1, len(data_part1), True)
```

```python
plot_data(data_part2, len(data_part2), True)
```

```python
plot_data(data_part3, len(data_part3), True)
```

```python
plot_data(data_part4, len(data_part4), False)
```
